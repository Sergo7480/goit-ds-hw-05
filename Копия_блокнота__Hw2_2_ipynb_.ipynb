Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
position_salary = df.groupby('Посада')['Зарплата на місяць'].agg(['min', 'max'])

avg_salary_by_position = df.groupby('Посада')['Зарплата на місяць'].mean()

def fill_avg_salary(row):
    try:
        return avg_salary_by_position.loc[row.name]
    except KeyError:
        return float('nan')

python_data = df.groupby('Посада').size()
python_data = python_data.to_frame(name='count')
python_data['avg'] = python_data.apply(fill_avg_salary, axis=1)
python_data['avg'] = python_data['avg'].round(0)

print(python_data)
```

Рівень зарплати в залежності від посади.

```python
import matplotlib.pyplot as plt
import seaborn as sns
position = (df.groupby('Посада')['Зарплата на місяць'].agg(['min', 'max']))
position.plot(kind='bar', figsize=(10, 6))
plt.title('Рівень зарплати в залежності від посади')
```

Рівень зарплати в залежності від доствіду

```python
experience = df.groupby(['Загальний досвід роботи', 'Посада'])['Зарплата на місяць'].agg(['min', 'max', 'mean'])
experience.plot(figsize=(10, 6))
plt.title('Рівень зарплати в залежності від доствіду')
plt.xticks(rotation=90)
```

Доля ринку програмістів за мовами програмування.

```python
import matplotlib.pyplot as plt

labels = df.groupby('Мова програмування')['Мова програмування'].count().index
sizes = df.groupby('Мова програмування')['Мова програмування'].count().values

plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Доля ринку програмістів за мовами програмування")
plt.axis('equal')
plt.show()
```
