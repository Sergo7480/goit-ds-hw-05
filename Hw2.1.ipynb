Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df.drop(df.index[-1], inplace=True)
print(df)
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
df.fillna(0, inplace=True)

df
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
df['2019'] = pd.to_numeric(df['2019'], errors='coerce')
df
```

```python
mean_2019 = df['2019'].mean()
mean_2019
```

```python
regions_above_average = df[df['2019'] > mean_2019]['Регіон']
print(regions_above_average)
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
df['2014'] = pd.to_numeric(df['2014'], errors='coerce')
df
```

```python
max_2014 = df['2014'].max()
print(max_2014)
```

```python
region_max = df.loc[df['2014'].idxmax(), 'Регіон']
region_max
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
import matplotlib.pyplot as plt
plt.bar(df['Регіон'], df['2019'])
plt.xticks(rotation=90)
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у 2019 році')
```

Народжуваність в 2000 році по регіонам України.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.barplot(x='2000', y='Регіон', data=df.sort_values('2000', ascending=False), palette='viridis')
plt.title('Уровень рождаемости по регионам Украины')
plt.xlabel('Рождаемость')
plt.ylabel('Регион')
plt.tight_layout()
plt.show()
```

Изменение народжуваності в Дніпропетровській області в порівнянні с 1990 і 2000 роками.

```python
import matplotlib.pyplot as plt
import pandas as pd

date = ['1990', '2000']
birth_rate = [16.6, 10.6]
plt.figure(figsize=(8, 6))
plt.title('Изменение народжуваності в Дніпропетровській області в порівнянні с 1990 і 2000 роками.')
plt.xlabel('Год')
plt.ylabel('Народжуваність')

plt.plot(date, birth_rate, marker='o', linestyle='-', color='b')
```
