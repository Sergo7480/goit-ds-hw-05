Тепер створіть новий датафрейм, який вміщатиме кількість книг для кожного з авторів (використовуйте функції groupby та agg, для підрахунку кількості використовуйте count). Не беріть усі стовпці, виберете тільки потрібні

```python
books_per_author = df.groupby('author')['name'].count().reset_index()
books_per_author.columns = ['author', 'book_count']
books_per_author
```

Відповідь: Якої розмірності вийшла таблиця? Відповідь: 248 x 2

Відповідь: Який автор має найбільше книг? Відповідь: Jeff Kinney


```python
books_per_author[books_per_author['book_count'] == books_per_author['book_count'].max()]
```

Відповідь: Скільки книг цього автора? Відповідь: 12

Тепер створіть другий датафрейм, який буде вміщати середній рейтинг для кожного автора (використовуйте функції groupby та agg, для підрахунку середнього значення використовуйте mean). Не беріть усі стовпці, виберете тільки потрібні

```python
average_author_rating = df.groupby('author')['user_rating'].mean().reset_index()
average_author_rating.columns = ['author', 'average_rating']
average_author_rating
```

Відповідь: У якого автора середній рейтинг мінімальний? Відповідь:Donna Tartt

```python
average_author_rating[average_author_rating['average_rating'] == average_author_rating['average_rating'].min()]
```

Відповідь: Який у цього автора середній рейтинг? Відповідь:3.9

З'єднайте останні два датафрейми так, щоб для кожного автора було видно кількість книг та середній рейтинг (Використовуйте функцію concat з параметром axis=1). Збережіть результат у змінну

```python
information_authors = pd.concat([books_per_author, average_author_rating['average_rating']], axis=1)
information_authors
```

Відсортуйте датафрейм за зростаючою кількістю книг та зростаючим рейтингом (використовуйте функцію sort_values)

```python
sort_by_rating = information_authors.sort_values(by=['book_count', 'average_rating'])
sort_by_rating
```

Рейтинг книжок

```python
import matplotlib.pyplot as plt
import seaborn as sns
rating=df['user_rating'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
sns.barplot(x=rating.index, y=rating.values)
plt.title('Распределение рейтингов книг')
plt.xlabel('Рейтинг')
plt.ylabel('Количество книг')
plt.show()
```

Продажі по авторам

```python
import seaborn as sns
import matplotlib.pyplot as plt

sale_by_author = df.groupby('author')['name'].count().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=sale_by_author.values, y=sale_by_author.index, palette="viridis")

plt.title("Топ-10 авторів за кількістю проданих книг")
plt.xlabel("Кількість проданих книг")
plt.ylabel("Автор")
plt.tight_layout()
plt.show()
```
